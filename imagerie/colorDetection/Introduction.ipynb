La détection d'objet dans une image s'appuie sur les caractéristiques des objets que l'on souhaitent voir. Il est important d'identifier les caractèristiques de ces objets avant de programmer un algorithme de détection.

Ce tutoriel permet de comprendre :
* Comment une image est coder
* Comment utiliser les différentes représentations de couleur d'une image
* Comment seuiller une image

```python
import cv2
import matplotlib.pyplot as plt
```

Le but de de tutoriel est de détecter les couleurs Rouge - Vert - Bleu de ce Rubix Cube

```python
imBGR = cv2.imread('../image/rubixCube.jpeg')
imRGB = cv2.cvtColor(imBGR, cv2.COLOR_BGR2RGB)
plt.imshow(imRGB)
plt.show()
```
