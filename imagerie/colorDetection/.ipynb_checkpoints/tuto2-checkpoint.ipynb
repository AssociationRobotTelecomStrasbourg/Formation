```python
import cv2
import matplotlib.pyplot as plt
import numpy as np 
```

# Trouver de vrais objets grâce à leur couleur
---
Si l'on regarde l'histogramme des channels RVB de l'objet suivant :

```python
im = cv2.cvtColor(cv2.imread('../image/redFlower.jpeg'), cv2.COLOR_BGR2RGB)
plt.imshow(im)

f,ax=plt.subplots(1,3,figsize=(20,5))
ax[0].hist(im[:,:,0].ravel(),255)
ax[0].set_title('Red')
ax[1].hist(im[:,:,1].ravel(),255)
ax[0].set_title('Green')
ax[2].hist(im[:,:,2].ravel(),255)
ax[0].set_title('Blue')

plt.show()
```
